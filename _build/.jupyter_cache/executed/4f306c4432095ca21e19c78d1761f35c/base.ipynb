```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/uit-bed-1304-h25/uit-bed-1304-h25.github.io/refs/heads/main/data/StudentPerformanceFactors.csv') 
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Anta at datasettet er lastet som df
sns.histplot(df['Exam_Score'], kde=True)
plt.title('Distribusjon av Karakterer')
plt.xlabel('Poengsum på eksamen')
plt.ylabel('Frekvens')
plt.show()
```

```python
# Lager en kopi av dataframen vår, og lager en ny kolonne
df_80 = df[df['Exam_Score'] >= 80]

# Samme som over
sns.histplot(df_80['Exam_Score'], kde=True)
plt.title('Distribusjon av Karakterer, for de over eller lik 80')
plt.xlabel('Poengsum på eksamen')
plt.ylabel('Frekvens')
plt.show()
```

```python
sns.boxplot(x='Gender', y='Exam_Score', data=df)
plt.title('Karakterfordeling etter Kjønn')
plt.xlabel('Kjønn')
plt.ylabel('Karakterer')
plt.show()
```

```python
#===================# 
## SeaBorn løsning ##
#===================#

sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Sammenheng mellom Studietid og Karakterer')
plt.xlabel('Studietid (timer per uke)')
plt.ylabel('Karakterer')
plt.show()
```

```python
#======================# 
## Matplotlib løsning ##
#======================#

# Lager et scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Hours_Studied'], df['Exam_Score'], color='blue', alpha=0.7)

# Legge til titler og etiketter
plt.title('Sammenheng mellom Studietid og Karakterer', fontsize = 14)
plt.xlabel('Studietid (timer per uke)', fontsize=12)
plt.ylabel('Karakterer', fontsize=12)

# Vise grafen
plt.show()
```

```python
# Velg numeriske kolonner
numerical_columns = ['Hours_Studied', 'Exam_Score', 'Sleep_Hours', 'Physical_Activity', 'Attendance', 'Previous_Scores']  # Legg til flere numeriske kolonner etter behov
corr_matrix = df[numerical_columns].corr()

# Lag et varmekart
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korrelasjon mellom ulike faktorer')
plt.show()
```

```python
import pandas as pd

df_bil = pd.read_csv('https://raw.githubusercontent.com/uit-bed-1304-h25/uit-bed-1304-h25.github.io/refs/heads/main/data/bilparken_ssb.csv') 
df_bil
```

```python
# Sjekk de første radene
df_bil.head()
```

```python
# Vis grunnleggende informasjon om datasettet
df_bil.info()
```

```python
# Beskriv statistikken til numeriske kolonner
df_bil.describe()
```

```python
# Sjekk for manglende verdier i datasettet
print(df_bil.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Omformer datasettet 
df_long = df_bil.melt(id_vars=['Tid', 'Bilmerker'], 
                      value_vars=['Under 4 år', '4 - 7 år', '8 - 11 år', '12 - 15 år', '16 - 20 år', 'Over 20 år',], 
                      var_name='Aldersgruppe', 
                      value_name='Antall_biler')

# Visualisering med Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=df_long, x='Bilmerker', y='Antall_biler', hue='Aldersgruppe', ci=None)
plt.xticks(rotation=90)
plt.title('Fordeling av biler i ulike aldersgrupper per bilmerke', fontsize=16)
plt.ylabel('Antall biler', fontsize=12)
plt.xlabel('Bilmerker', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Sorter bilmerkene etter prosentandel vrakede biler
df_sorted = df_bil.sort_values(by='Personbiler vraket mot pant (prosent av bestanden)', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_sorted, x='Bilmerker', y='Personbiler vraket mot pant (prosent av bestanden)', palette='viridis')
plt.xticks(rotation=90)
plt.title('Bilmerker med høyest prosentandel vrakede biler', fontsize=16)
plt.ylabel('Prosentandel vrakede biler', fontsize=12)
plt.xlabel('Bilmerker', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Velg numeriske kolonner for korrelasjon
corr_columns = ['Gjennomsnittsalder ved vraking (år)', 
                'Personbilbestand 31. desember', 
                'Personbiler vraket mot pant', 
                'Personbiler vraket mot pant (prosent av bestanden)']

# Beregn korrelasjonen
corr_matrix = df_bil[corr_columns].corr()

# Plot korrelasjonsmatrise med Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Korrelasjon mellom numeriske variabler', fontsize=16)
plt.tight_layout()
plt.show()
```
